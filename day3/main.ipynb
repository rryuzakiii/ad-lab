```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.read_csv("./data.csv")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scaler = StandardScaler()
```

```python
train, test = train_test_split(df, test_size=0.2)
```

```python
X_train = train[['Socioeconomic Score',
 'Study Hours',
 'Sleep Hours',
 'Attendance (%)']]

X_test = test[['Socioeconomic Score',
 'Study Hours',
 'Sleep Hours',
 'Attendance (%)']]
```

```python
y_train = train[['Grades']]
y_test = test[['Grades']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression(n_jobs=5)
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
mae = mean_absolute_error(y_pred=y_pred, y_true=y_test)
mse = mean_squared_error(y_pred=y_pred, y_true=y_test)
r2 = r2_score(y_pred=y_pred, y_true=y_test)
```

```python
print(f"R2 Score: {r2}")
print(f"MSE: {mse}")
print(f"MAE: {mae}")
```

Logistic Regression

```python
df = pd.read_csv("./diabetes.csv")
```

```python
train1, test1 = train_test_split(df, test_size=0.2)
```

```python
X_train1 = train1[['Pregnancies',
 'Glucose',
 'BloodPressure',
 'SkinThickness',
 'Insulin',
 'BMI',
 'DiabetesPedigreeFunction',
 'Age']]

X_test1 = test1[['Pregnancies',
 'Glucose',
 'BloodPressure',
 'SkinThickness',
 'Insulin',
 'BMI',
 'DiabetesPedigreeFunction',
 'Age']]
```

```python
print(train1.columns)
```

```python
y_train1 = train1[['Outcome']]
y_test1 = test1[['Outcome']]
```

```python
y_train1 = y_train1.values.ravel()
y_test1 = y_test1.values.ravel()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
model = LogisticRegression(n_jobs=5, max_iter=500)
model.fit(X_train1, y_train1)
```

```python
y_pred1 = model.predict(X_test1)
accuracy = accuracy_score(y_test1, y_pred1)
print(f"Accuracy: {accuracy:.2f}")
```

```python
cm = confusion_matrix(y_test1, y_pred1)
print("Confusion Matrix:")
print(cm)
```

```python
report = classification_report(y_test1, y_pred1)
print("Classification Report:")
print(report)
```
